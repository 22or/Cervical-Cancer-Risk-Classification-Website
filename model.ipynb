```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# get data

```python
df = pd.read_csv('kag_risk_factors_cervical_cancer.csv',na_values = '?')
df
```

```python
df.info()
```

#  clean data

### drop time since first/last std diagnosis columns

```python
df = df.drop('STDs: Time since first diagnosis',axis = 1)
df = df.drop('STDs: Time since last diagnosis',axis = 1)
df
```

### impute missing values as mean

```python
df = df.fillna(df.mean())
df
```

```python
# split independent variables and dependent variable

vars = df.drop('Biopsy',axis = 1)
result = df['Biopsy']
```

# prepare data for training/testing

### split data into training data and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
vars_train,vars_test,result_train,result_test = train_test_split(vars,result,test_size = 0.2,random_state = 10)
```

```python
vars_train
```

```python
vars_test
```

### standardize scale of data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
vars_train_scaled = scaler.fit_transform(vars_train)
vars_test_scaled = scaler.transform(vars_test)
```

```python
vars_train_scaled
```

```python
vars_test_scaled
```

# train model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state = 10).fit(vars_train_scaled,result_train)
```

```python
print(str(model.score(vars_train_scaled,result_train)) + " accurate on training data")
```

```python
print(str(model.score(vars_test_scaled,result_test)) + " accurate on testing data")
```

# evaluate model

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
result_predict = model.predict(vars_test)
print(classification_report(result_test,result_predict))
```

```python
plt.figure(figsize=(10,10))
cm = confusion_matrix(result_predict,result_test) 
sns.heatmap(cm,annot = True,fmt = '.2f')
plt.ylabel('predicted')
plt.xlabel('actual')
```

# train model with xgboost

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBClassifier(learning_rate = 0.1, max_depth = 50, n_estimators = 1000).fit(vars_train,result_train)
```

```python
print(str(xgb_model.score(vars_train_scaled,result_train)) + " accurate on training data")
```

```python
print(str(xgb_model.score(vars_test_scaled,result_test)) + " accurate on testing data")
```

# evaluate new model

```python
xgb_result_predict = xgb_model.predict(vars_test)
print(classification_report(result_test,result_predict))
```

```python
plt.figure(figsize=(10,10))
cm = confusion_matrix(xgb_result_predict,result_test) 
sns.heatmap(cm,annot = True,fmt = '.2f')
plt.ylabel('predicted')
plt.xlabel('actual')
```
